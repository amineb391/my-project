# Visualisation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


Data.hist(bins=15,figsize=(15,10),layout=(4,3))
plt.suptitle('Histogramme des variables')
plt.show()
```

```python
Data.plot(kind='density',layout=(3,3),subplots=True,sharex=False,sharey=False,figsize=(15,10))
plt.show()
```

```python
plt.figure(figsize=(15,10))
corr=Data.select_dtypes(include='number').corr()
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Matrice de correlation')
plt.show()
```

```python
Data.plot(kind='box',layout=(3,3),subplots=True,sharex=False,sharey=False,figsize=(15,10))
plt.title('Boite a moustache')
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(Data,figsize=(15,10),c='g')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(Data,hue='Channel')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

Data=read_csv(fch,names=col)
corr=Data.select_dtypes(include='number').corr()
sums = corr.sum()
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(aspect="equal"))
colors = plt.cm.tab20c.colors
wedges, texts, autotexts = ax.pie(sums, labels=sums.index,
autopct='%1.1f%%', startangle=90, colors=colors)
ax.set_title("Histogramme en cercle des sommes des corrélations")
plt.setp(autotexts, size=10, weight="bold", color="white")
plt.setp(texts, size=12)
plt.show()
```
